# Series
Она предлставляет из себя объект, похожий на одномерный массив, но отчилительной чертой являет наличие индексов. Индекс находится слева, а сам элемент справа.

Синтаксис создания:
```pandas.Series(input_data, index, data_type)```
- input_data: ввод в виде списка, контсанты, масива NumPy, Dict и т.д.
- index: значение индексов.
- data_type (опционально): тип данных.

````python
import pandas as pd
````

````python
a = pd.Series([4, 7, 6, 3, 9],
              index=['one', 'two', 'three', 'four', 'five'])
a
````

````python
a = pd.Series([4, 7, 6, 3, 9])
a
````

````python
a.index
````

````python
a.values
````

````python
a[0]
````

````python
a[1]
````

````python
a['one']
````

# DataFrame
Объект DataFrame является табличной структурой данных. В любой таблице всегда присутствуют строки и столбцы. При этом в столбцах можно хранить данные разных типов данных. Столбцами в объекте DataFrame выступают объект Series, строки которых являются их элементами.

Синтацсис создания: ```pandas.DataFrame(input_data, index)```
- input_data: ввод в виде Dict, 2D массива NumPy, Series и т.д.
- index: значения индексов.

````python
df = pd.DataFrame({
    'Age': [46, 37, 44, 42, 42],
    'Country': ['Spain', 'Spain', 'Germany', 'Germany', 'France'],
    'Gender': ['Female', 'Female', 'Male', 'Male', 'Male']
})
df
````

````python
df['Age']
````

````python
df.Country
````

````python
df[['Country', 'Age']]
````

````python
df.columns
````

````python
df.index
````

````python
df = pd.DataFrame({
    'Age': [46, 37, 44, 42, 42],
    'Country': ['Spain', 'Spain', 'Germany', 'Germany', 'France'],
    'Gender': ['Female', 'Female', 'Male', 'Male', 'Male']
}, index=[5,4,6,3,2])
df
````

# Фильтрация
Фильтрация в pandas основывается на булевых масках.

> Булевая маска — бинарные данные, которые используются для выбора определенных объектов из структуры данных.

````python
df['Gender'] == 'Male'
````

````python
male = df[df['Gender'] == 'Male']
male
````

# Логические И
При операторе & нужно, чтобы выполнялось два условия одновременно:

````python
df[(df['Gender'] == 'Female') & (df['NumOfProducts'] >= 3)]
````

````python
(df['Gender'] == 'Female') & (df['NumOfProducts'] >= 3)
````

# Логические ИЛИ
При операторе | нужно, чтобы выполнялось хотя бы одно условие:

````python
(df['HasCrCard']) | (df['NumOfProducts'] >= 3)
````

````python
df[(df['HasCrCard']) | (df['NumOfProducts'] >= 3)]
````

# Логические НЕ
При операторе ~ булевая маска обращается: True меняется на False и наоборот:

````python
df[~(df['Geography'] == 'Spain')]
````

````python
df[df['Geography'].isin(['France', 'Germany'])]
````
