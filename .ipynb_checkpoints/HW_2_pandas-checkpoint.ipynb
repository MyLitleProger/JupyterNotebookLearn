### 4.1 Сколько в среднем стоят дома, у которых 2 спальни?

### 4.2 Какая в среднем общая площадь домов, у которых стоимость больше 600 000?

20454

### 4.3 Как много домов коснулся ремонт?

### 4.4 Насколько в среднем стоимость домов с оценкой grade домов выше 10 отличается от стоимости домов с оценкой grade меньше 4?

````python
df_grade_10 = df[df['grade'] >= 10][['price', 'grade']]
df_grade_10
````

````python
df_grade_4 = df[df['grade'] <= 4][['price','grade']]
df_grade_4
````

````python
df_grade_10.mean()
````

````python
df_grade_4.mean()
````

````python
df_grade_10.mean() - df_grade_4.mean()
````

стоимость различается:

```1 257 985 - 211 395 = 1 046 589```

### 5.1 Выберите дом клиенту
Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?

````python
df_home = df[(df['waterfront']>0) & (df['bathrooms']>=3)&(df['sqft_basement']>0)][['waterfront','bathrooms','sqft_basement']]
df_home
````

````python
df_home.count()
````

У клиента есть 41 вариант

### 5.2 Выберите дом клиенту
Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В какой ценовом диапазоне будут дома?

````python
df['view'].max()
````

````python
df[((df['view']==4) | (df['waterfront'] > 0)) & (df['yr_built']>=1980)][['price']].describe()
````
